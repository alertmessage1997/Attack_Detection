# Splitting the dataset into features and target

```python
x = df.drop(['outcome','attack'], axis = 1)
Y = df['attack']
```

# Feature Selection -> Chi Square Test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# applying selectKbest to extract top 20 best features
bestfeatures = SelectKBest(score_func = chi2, k=20)
X_kbest = bestfeatures.fit(x,Y)
```

```python
dfscores = pd.DataFrame(X_kbest.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
# concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis =1)
featureScores.columns = ['Specs', 'Score']
```

```python
featureScores
```

```python
feature_sel = featureScores.nlargest(20, 'Score')
feature_sel
```

# Model Building

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.layers import GRU
from keras.optimizers import Adam
```

```python
model = Sequential()  # create a new sequential model using Sequential API

# multi layer of CNN
model.add(Conv1D(filters=128, kernel_size=3, activation='relu', input_shape=(20,1))) # creates a new convolutional 1D layer
model.add(Conv1D(filters=64, kernel_size=3, activation='relu')) # creates another convolutional 1D layer
model.add(Conv1D(filters=32, kernel_size=3, activation='relu')) # creates another convolutional 1D layer
model.add(Conv1D(filters=16, kernel_size=3, activation='relu')) # creates another convolutional 1D layer
model.add(Conv1D(filters=16, kernel_size=3, activation='relu')) # creates another convolutional 1D layer

model.add(MaxPooling1D(pool_size=2)) # creates a max pooling layer

# first GRU layer
model.add(GRU(units=64, return_sequences=True, activation='relu')) # creates a GRU layer
model.add(Dropout(0.5))  # adds a dropout layer to handle over-fitting.

# second GRU layer
model.add(GRU(units=32, return_sequences=True, activation='relu')) # creates another GRU Layer
model.add(Dropout(0.5))  # adds a dropout layer to handle over-fitting.

model.add(Flatten()) # flattens the input
model.add(Dense(100, activation='relu')) # adds a dense layer with relu activation function.
model.add(Dense(1, activation='sigmoid')) # adds final dense layer with sigmoid activation function.
```

```python
opt=Adam(learning_rate=0.00001)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=128, epochs=10, validation_data=(x_test, y_test))
```

```python
loss, accuracy = model.evaluate(x_test, y_test, batch_size=128)
print(accuracy)
```
